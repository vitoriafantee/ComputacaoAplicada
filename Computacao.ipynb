```python
# Preços de Gasolina A sem tributos, à vista, por vigência (R$/m3)
# https://precos.petrobras.com.br/
# a biblioteca pandas é usada para manipulação e análise de dados
import pandas as pd
# lendo o arquivo em excel
tabela = pd.read_excel("/content/drive/MyDrive/trabalho_computacao/valorgasolina.xlsx")
display(tabela)
```

```python
#para agrupar por cidade os valores
tab_local = tabela.groupby('LOCAL').sum()
display(tab_local)
```

```python
# para selecionar um mês específico e ordenar os valores totais para cada cidade
tab_local = tab_local[['jun']].sort_values(by='jun', ascending=False)
display(tab_local)
```

```python
#gráfico de barras com os preços agrupados por cidade, independente da modalidade de venda
import plotly.express as px
grafico = px.bar(tab_local, x=tab_local.index, y='jun')
grafico.show()
```
